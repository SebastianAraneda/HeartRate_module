```python
import xml.etree.ElementTree as ET
camTree = ET.parse('datasets/LGI-PPGI/cpi/cpi_talk/cv_camera_sensor_timer_stream_handler.xml')
cmsTree = ET.parse('datasets/LGI-PPGI/cpi/cpi_talk/cms50_stream_handler.xml')

camFrames = [] #25 fps
cmsFrames = [] #60 fps

myroot = camTree.getroot()
for i in range(len(myroot)):
    camFrames.append([
        float(myroot[i][0].text),
        float(myroot[i][1].text)])

myroot = cmsTree.getroot()
for i in range(len(myroot)):
    cmsFrames.append([int(myroot[i][0].text),int(myroot[i][1].text),int(myroot[i][2].text)])
```

```python
import matplotlib.pyplot as plt
x= [] 
cap = []
hr = []
for i in range(len(cmsFrames)):
#for i in range(2000):    
    x.append(cmsFrames[i][0]) # frame
    hr.append(cmsFrames[i][1]) # heart rate
    cap.append(cmsFrames[i][2]) # data captured

caphr = []

for i in range(len(cmsFrames)):
    caphr.append(( int(hr[i])**2 + int(cap[i])**2) )

plt.plot(x, hr, label = "bps")
plt.plot(x, cap, label = "data", linestyle='dashed', markerfacecolor='blue')

plt.xlabel('Frame')
plt.ylabel('Intensity')
plt.title('Oximeter data')
plt.show()


```
