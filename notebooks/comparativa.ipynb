```python
from pyVHR.analysis.pipeline import Pipeline
import matplotlib.pyplot as plt

times, bpms, uncertainties, tittles, aproach, roi_met = [],[],[],[],[],[]
rm = ('convexhull' ,'faceparsing')
ra = ('patches', 'holistic')
```
